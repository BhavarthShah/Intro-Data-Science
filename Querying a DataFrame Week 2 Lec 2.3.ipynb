# Querying a DataFrame 
Week 2 Lec 2.3

```python
import pandas as pd
import numpy as np
```

```python
#boolean masking for efficient querying in numpy and pandas

df = pd.read_csv("datasets/Admission_Predict.csv", index_col = 0)

df.columns = [x.lower().strip() for x in df.columns]

df.head()
```

```python
#broadcasting comparison for the data that we want only in our set

admit_mask = df['chance of admit'] > 0.7
```

```python
admit_mask
```

```python
#using boolean mask to lean it on our dataframe

df.where(admit_mask).head()
```

```python
#it shows NaN value where the value is false to drop those we use .dropna()

df.where(admit_mask).dropna()
```

```python
#combine .where() & .dropna() -- we use indexing operator 

df[df['chance of admit'] > 0.7].head()
```

# Indexing Operator 3 uses

## 1. Call String Parameter

```python
#use the string name to find the parameter

df["gre score"]
```

## 2. List of Columns

```python
df[["gre score", "chance of admit"]]
```

## IMPORTANT 3. Send a Boolean Mask -- Also include filter

```python
df[df["gre score"] > 320]

#only gives the value higher than 320 gre score
```

## Multiple boolean masks

```python
(df['chance of admit']>0.7) & (df['gre score ']>320)

#this will be an error since gre score space is invalid
```

```python
(df['chance of admit']>0.7) & (df['gre score']>320)

#returns true for both >0.7 and gre >320
```

```python
df[(df['chance of admit']>0.7) & (df['gre score']>320)]

#only displaces one with both conditions meeting
```

```python
#another format for the same task

df[df['chance of admit'].gt(0.7).lt(0.9)]
```

```python
df[df['chance of admit'].gt(0.7) | df['gre score'].gt(320)]

#OR operator
```
